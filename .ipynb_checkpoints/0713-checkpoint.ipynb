```python
#page 1 & 2
import requests
from bs4 import BeautifulSoup
i=1
while (i<=2):
    nextlink = "https://www.104.com.tw/cust/list/index/?page="+str(i)+"&order=1&mode=s&jobsource=checkc"
    i=i+1
    res = requests.get(nextlink)
    soup = BeautifulSoup(res.text)
    for entry in soup.select(".items"):
        print(entry.select('h1')[0].find('a').text)
        print(entry.select(".data")[0].text)
        print(entry.select(".data")[1].text)
        print("=========================================")
```
